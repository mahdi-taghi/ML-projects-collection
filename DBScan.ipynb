```python
df_95 = pd.read_csv('newfile_gorgan_95.csv', encoding='utf-8')
df_96 = pd.read_csv('newfile_gorgan_96.csv', encoding='utf-8')
df_97 = pd.read_csv('newfile_gorgan_97.csv', encoding='utf-8')
df_98 = pd.read_csv('newfile_gorgan_98.csv', encoding='utf-8')
df_99 = pd.read_csv('newfile_gorgan_99.csv', encoding='utf-8')
```

```python
df_95 = df_95.drop('Unnamed: 0', axis=1)
df_96 = df_96.drop('Unnamed: 0', axis=1)
df_97 = df_97.drop('Unnamed: 0', axis=1)
df_98 = df_98.drop('Unnamed: 0', axis=1)
df_99 = df_99.drop('Unnamed: 0', axis=1)
```

```python
all_data = pd.concat([df_95, df_96, df_97, df_98, df_99], ignore_index=True)
```

```python
all_data = all_data[all_data['تعویض کنتور'].isna()]
```

```python
data_maskoni = all_data[all_data['کاربری'] == 'مسكوني']
data_maskoni
```

```python
#shomordane tedad eshterakha
counts = data_maskoni.groupby('کد اشتراک').size().reset_index(name='تعداد ماه‌های قرائت')

#braye moghayese balaye 48 mah
threshold = 0.75 * 48

#balaye 75 darsad 
more_than_75_percent = counts[counts['تعداد ماه‌های قرائت'] > threshold]
#paiin ya barabare 75
less_than_or_equal_75_percent = counts[counts['تعداد ماه‌های قرائت'] <= threshold]


#code eshterak ro be onvame index dar nazar migirim
data_maskoni_indexed = data_maskoni.set_index('کد اشتراک')

#az ham do horoh ro joda mikonim va reset index mikonim ta data moratab bshe
data_more_than_75_percent = data_maskoni_indexed.loc[more_than_75_percent['کد اشتراک']]
data_less_than_or_equal_75_percent = data_maskoni_indexed.loc[less_than_or_equal_75_percent['کد اشتراک']]

data_more_than_75_percent = data_more_than_75_percent.reset_index()
data_less_than_or_equal_75_percent = data_less_than_or_equal_75_percent.reset_index()

print(f"تعداد مشترکین با قرائت در بیش از 75% ماه‌ها در دیتای اصلی: {data_more_than_75_percent.shape[0]}")
print(f"تعداد مشترکین با قرائت در کمتر از یا مساوی با 75% ماه‌ها در دیتای اصلی: {data_less_than_or_equal_75_percent.shape[0]}")
```

```python
balaye_75=(len(data_more_than_75_percent)*100)/len(data_maskoni)
zire_75=(len(data_less_than_or_equal_75_percent)*100)/len(data_maskoni)
print('moshtarakin ba gherat zire 75%: ',round(zire_75,2))
print('moshtarakin ba gherat balaye 75%: ',round(balaye_75,2))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x= scaler.fit_transform(data_maskoni_1)
```

```python
eps=[2,1,0.5,0.1]
for i in eps:
    from sklearn.cluster import DBSCAN
    dbscan=DBSCAN(eps=i,min_samples=2)
    dbscan.fit(x)
    labels=dbscan.labels_
    n_clusters_=len(set(labels))-(1 if 10 in labels else 0)
    unique_labels=set(labels)

    noise_points = x[dbscan.labels_ == -1]
    print('eps ',i,' : ',len(noise_points))
```

```python
min_samples=[5,10,20,40,60,80,100]
for i in min_samples:
    from sklearn.cluster import DBSCAN
    dbscan=DBSCAN(eps=0.1,min_samples=i)
    dbscan.fit(x)
    labels=dbscan.labels_
    n_clusters_=len(set(labels))-(1 if 10 in labels else 0)
    unique_labels=set(labels)

    noise_points = x[dbscan.labels_ == -1]
    print('min sample ',i,' : ',len(noise_points))
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.1,min_samples=100)
dbscan.fit(x)
labels=dbscan.labels_
n_clusters_=len(set(labels))-(1 if 10 in labels else 0)
unique_labels=set(labels)
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,3],c=labels)
plt.show()
noise_points = x[dbscan.labels_ == -1]
len(noise_points)
```

```python
noise_points = x[dbscan.labels_ == -1]
original_noise_points = scaler.inverse_transform(noise_points)
df=pd.DataFrame(original_noise_points)
df.columns = ['مصرف', 'مبلغ', 'واحد', 'ظرفیت']
shared_indices = data_maskoni_1.apply(lambda row: tuple(row) in [tuple(r) for r in df.values], axis=1)
shared_df1 = data_maskoni_1[shared_indices]
share=pd.DataFrame(shared_df1)
columns_of_interest = ['مصرف', 'مبلغ','واحد','ظرفیت']
eshterak = data_maskoni[columns_of_interest].apply(lambda row: tuple(row) in [tuple(r) for r in share[columns_of_interest].values], axis=1)

esh1 = data_maskoni[eshterak]
sh1=pd.DataFrame(esh1['کد اشتراک'])
sh1=sh1.drop_duplicates()
```
