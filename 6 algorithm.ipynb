```python
#SOM
from minisom import MiniSom
feature_columns = [col for col in data.columns if col not in ['sex', 'age', 'experience', 'education']]
data_clean = data[feature_columns].dropna()
data_np = data_clean.values


som = MiniSom(6, 6, len(feature_columns), sigma=0.5, learning_rate=0.5) 
som.train_random(data_np, 100)

from pylab import bone, pcolor, colorbar, plot, show

bone()
pcolor(som.distance_map().T)
colorbar()
show()
```

```python
from sklearn.manifold import TSNE

win_map = som.win_map(data_clean.values)
clustered = pd.DataFrame(columns=data_clean.columns)
for x, t in win_map.items():
    df = pd.DataFrame(t, columns=data_clean.columns)
    df['Cluster'] = f'Cluster({x[0]},{x[1]})'
    clustered = pd.concat([clustered, df])

tsne = TSNE(n_components=2, random_state=0)
tsne_results = tsne.fit_transform(data_clean.values)

clustered['t-SNE-2d-one'] = tsne_results[:,0]
clustered['t-SNE-2d-two'] = tsne_results[:,1]


for cluster in clustered['Cluster'].unique():
    print(f"Analyzing {cluster}")
    cluster_data = clustered[clustered['Cluster'] == cluster]
```

```python
from sklearn.decomposition import FactorAnalysis
 
factor_analysis = FactorAnalysis(n_components=2) 


fa_leadership = factor_analysis.fit_transform(data_clean)


fa_leadership_df = pd.DataFrame(fa_leadership, columns=['Leadership Factor 1', '2'])

print(fa_leadership_df.head())
```

```python
from sklearn.cluster import KMeans
ys=[]
for k in np.arange(1,14):
    kmn=KMeans(n_clusters=k)
    kmn.fit(data)
    ys.append(kmn.inertia_)
plt.plot(np.arange(1,14),ys,'o-')
plt.xlabel('number of clusters')
plt.ylabel('inertia')
plt.show()
```

```python
#unsupervise
kmn=KMeans(n_clusters=8)
kmn.fit(data)
labels=kmn.predict(data)
center=kmn.cluster_centers_
plt.scatter(data.iloc[:,4],data.iloc[:,51],c=labels)
plt.scatter(center[:,4],center[:,51],marker='x',s=150,color='r')
plt.show()
```

```python
#unsupervize DBScan
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=1,min_samples=1)
dbscan.fit(data)
labels=dbscan.labels_
n_clusters_=len(set(labels))-(1 if 10 in labels else 0)
unique_labels=set(labels)
plt.scatter(data.iloc[:,4],data.iloc[:,51],c=labels)
plt.show()
```

```python
y=data['education']
x=data
x.drop('education',inplace=True,axis=1)
x
```

```python
#supervise
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

log_reg = LogisticRegression(max_iter=1000)

log_reg.fit(x_train, y_train)


from sklearn.metrics import log_loss

prob_predictions = log_reg.predict_proba(x_test)
loss = log_loss(y_test, prob_predictions)

print("Log Loss:", loss)


predictions_lr = log_reg.predict(x_test)
print(classification_report(y_test, predictions_lr))
print(confusion_matrix(y_test, predictions_lr))
```

```python
#MLP
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(100,80,50), max_iter=5000)

mlp.fit(x_train, y_train)



prob_predictions = mlp.predict_proba(x_test)

loss = log_loss(y_test, prob_predictions)

print("Log Loss:", loss)



predictions_mlp = mlp.predict(x_test)
print(classification_report(y_test, predictions_mlp))
print(confusion_matrix(y_test, predictions_mlp))
```
