```python
data[['sex','age','Medu','Fedu','guardian','traveltime','studytime','failures','schoolsup','paid','activities','nursery','higher','internet','romantic','famrel','freetime','goout','Dalc','Walc','health','absences' ]]=data[['sex','age','Medu','Fedu','guardian','traveltime','studytime','failures','schoolsup','paid','activities','nursery','higher','internet','romantic','famrel','freetime','goout','Dalc','Walc','health','absences' ]].apply(pd.to_numeric,errors='coerce')
data=data.dropna()
data=data.reset_index(drop=True)
print(data.size)
data.head(5)
```

```python
from sklearn.preprocessing import MinMaxScaler
featureset=data[['sex','age','Medu','Fedu','guardian','traveltime','studytime','failures','schoolsup','paid','activities','nursery','higher','internet','romantic','famrel','freetime','goout','Dalc','Walc','health','absences' ]]
y=featureset.values
min_max_scaler=MinMaxScaler()
feature_mtx=min_max_scaler.fit_transform(y)
feature_mtx[0:5]
```

```python
import scipy 
f=feature_mtx.shape[0]
d=scipy.zeros([f,f])
for i in range(f):
    for j in range(f):
        d[i,j]=scipy.spatial.distance.euclidean(feature_mtx[i],feature_mtx[j])
d
```

```python
import pylab
import scipy.cluster.hierarchy 
z=hierarchy.linkage(d,'complete')
z
```

```python
fig=pylab.figure(figsize=(18,60))
def rrr(id):
    return '[%s %s %s]'%(data['sex'][id],data['studytime'][id],int(float(data['internet'][id])))
dendro=hierarchy.dendrogram(z,leaf_label_func=rrr,leaf_rotation=0,leaf_font_size=12,orientation='right')
```
